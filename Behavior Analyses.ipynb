### The below graphs of cumulative distribution function show important pattern that will branch out the project to classification machine learning (ML) models. These models will be logestic regression, KNN, SVC, kernel SVC, and Naive Bayes.
The 0 excretion represents more than 50% of total excretion. The 0 phenomenon looks not coincidence as it happened in all days (Day1, Day2, and Day3) with almost the same percent. Based on this pattern, we converted the 'Excretion' parameter into binary outcome, 0 vs not 0. The ML classifiers will handle this in separated sub-projects. Please find these sub-projects following the link https://github.com/al-naimi/Voles/blob/master/Excretions-Binary%20outcome.ipynb

# Below is Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset.head(2)
```

```python
# creating matrix of features to identify the dependent factor vs independent factors (parameters)
X = dataset.iloc[:, 4].values  
y = dataset.iloc[:, 6].values 
```

```python
type (X)
```

```python
type (y)
```

```python
X = X.reshape((194,1)) # Use it to avoid error message (ValueError: Found arrays with inconsistent numbers of samples: [  1 999])
# regressor.fit(np.transpose(np.matrix (X_train)), np.transpose(np.matrix (y_train))) # this of value error
# regressor.fit (X_train.reshape(len(X_train), 1), y_train)
```

```python
type (X)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0) # you can ommit random_state. It 
# keeps the same random sample for all sampling trials OR you can choose different value.
```

```python
# Feature Scaling. This for bringing large and small values of different parameters on the same scale.
# No need for feature scaling as linear regression library will do it for us
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
# Fitting classifier to the Training set & Predicting the Test set results
# Fitting Simple Linear Regression to the Training set. Creating the regression class (classifier)
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit (X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
# visualizing the training test results
plt.scatter (X_train, y_train, color = 'red')
plt.plot (X_train, regressor.predict (X_train), color = 'blue')
plt.title ('Excretions correlation')
plt.xlabel ('Day 1')
plt.ylabel ('Day 2')
plt.show ()
```

```python
# visualizing the Test test results
plt.scatter (X_test, y_test, color = 'red')
plt.plot (X_train, regressor.predict (X_train), color = 'blue')
plt.title ('Excretions correlation')
plt.xlabel ('Day 1')
plt.ylabel ('Day 2')
plt.show ()
```

### The conclusion
1. There is positive linear correlation between Day1 vs Day2 and Day1 vs Day3
2. The interpretation is that the voles who showed high excretory output in Day1, also showed that in Day2 and Day3 and
vice versa. This might mean that the stressed animal will continue to be so throughout the 3 days of the experiment.
